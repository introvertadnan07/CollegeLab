```python
# 1. create a tuple
my_tuple = (10, 20,30,40,50)

# 2 . print origibal tuple
print("Orginal Tuple : ", my_tuple)

# 3. Access element
print("First element : ", my_tuple[0])
print("Last element : ", my_tuple[-1])

# 4. Slicing 
print("Middle element: ", my_tuple[1:4])

# 5. Tuple concatenation
tuple2 = (60, 70)
combined = my_tuple + tuple2
print("After Concatenation: ", combined)

# 6. Tuple Repetition
repeated = my_tuple * 2
print("After repettion : ", repeated)

# 7. check membership
print("1s 30 in tuple ?: ", 30 in my_tuple)
print("1s 100 in tuple ?: ", 100 in my_tuple)

# 8. Tuple Function
print("Length of tuple:", len(my_tuple))
print("Maximum value: ", max(my_tuple))
print("Minimum value: ", min(my_tuple))
print("Count of 20: ",my_tuple.count(20))
print("Index of 40: ", my_tuple.index(40))
```
