```python
# print Even number from 1 to 10 using for loop
print("Even number from 1 to 10: ")
for i in range (1, 11):
    if i % 2 == 0:
        print(i)
```

```python
# Count from 1 to 5 using while Loop
i = 1
print("Counting from 1 to 5: ")
while i <= 5:
    print(i)
    i += 1
```

```python
# Use to exit break loop early
print("Break loop at 3:")

for i in range(1, 6):  
    if i == 3:
        break           
    print(i)
```
