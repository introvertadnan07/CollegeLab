```python
# Writing to a text file
file = open("sample.txt", "w")   # Use .txt, not .text
file.write("Hello this is a text file.\nWelcome to Python file handling.")
file.close()

# Reading from the text file
file = open("sample.txt", "r")
content = file.read()
print("Content of text file:")
print(content)
file.close()


# Binary file read and write using pickle
import pickle

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print("Name:", self.name)
        print("Marks:", self.marks)


# Create object
s1 = Student("Ravi", 91)

# Write object to binary file
with open("student.data", "wb") as f:
    pickle.dump(s1, f)

# Read object from binary file
with open("student.data", "rb") as f:
    s2 = pickle.load(f)

print("Data from file:")
s2.display()
```
