```python
# Define a sample string
text = "Python for Data Science"
# print original string
print("Original string.", text)

# length of the string 
print("Length:", len(text))

# String slicing
print("First 6 characters: ", text[:6])
print("last 7 characters: ", text[-7])

# convert to uppercase and lowercase
print("Uppercase: ", text.upper())
print("Uppercase: ", text.lower())

# Replace a word
print("Replace","python with","Machine Learning:", text.replace("Python", "Machine Learning"))

# Count occurrence of character
print("Count of 'a': ", text.count('a'))

# find poistion of a word
print("position of Data : ", text.find("Data"))

text = "Python"
newtext = text + " is interesting"
print("After concatenation:", newtext)

# 9. Split the string
words = text.split()
print("Split into words : ", words)

# Join words back with hypen

words = ["Python", "is", "fun"]

joined = "-".join(words)
print("Joined with '-':", joined)
```
