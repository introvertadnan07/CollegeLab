```python
# 1. Create dictinory
student = {
    "name": "Adnan",
    "age": 22,
    "branch": "CSE",
    "marks": 98
}

print("Original Dictionary:", student)

# 2. Access values using keys
print("Name:", student.get("name"))
print("Age:", student.get("age"))

# 3. Update value
student["marks"] = 92
student["age"] = 23
print("After updating marks and age : ", student)

# 4. Add new key_valu pair
student["college"] = "Avit"
print("After adding college: ", student)

# 5. Delete a key
del student["branch"]
print("After deleting branch: ", student)

# 6. Dictionary Methods
print("Keys: ", student.keys())
print("values: ", student.values())
print("Items: ", student.items())

# uSE OPO() AND CLEAR
student.pop("marks: ", student)
print("After popping marks: ", student)

student.clear()
print("After clearning dictionary: ", student)
```
