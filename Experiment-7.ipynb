```python
# Define a class
class Student:
    # Constructor to initialize object
    def __init__(self, name, age, marks):
        self.name = name
        self.age = age
        self.marks = marks

    # Method to display student details
    def display(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("Marks:", self.marks)

    # Method to check if student passed
    def has_passed(self):
        if self.marks >= 40:
            print("Status: Passed")
        else:
            print("Status: Failed")


# Create objects of the class
student1 = Student("Adnan", 22, 87)
student2 = Student("Aman", 21, 32)

# Use object methods
print("Student 1:")
student1.display()
student1.has_passed()

print("\nStudent 2:")
student2.display()
student2.has_passed()
```
