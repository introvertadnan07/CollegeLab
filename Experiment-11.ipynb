```python
import pandas as pd

df1 = pd.DataFrame({
    'ID': [1, 2],
    'Name': ['Alice', 'Bob']
})

df2 = pd.DataFrame({
    'ID': [3, 4],
    'Name': ['Afzal', 'Bushra']
})

# 1. Concatenation
combined_df = pd.concat([df1, df2])
print("Vertical concatenation:\n", combined_df)


# Create two dataframes for merging
student_info = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Aman']
})

marks_info = pd.DataFrame({
    'ID': [1, 2, 3],
    'Marks': [87, 83, 91]
})

# 2. Merge on common key (ID)
merged_df = pd.merge(student_info, marks_info, on='ID')
print("\nMerged DataFrame:\n", merged_df)


# Create two dataframes for index-based join
left = pd.DataFrame({'Math': [88, 98]},
                    index=['Alice', 'Bob'])
right = pd.DataFrame({'Science': [85, 90]},
                     index=['Alice', 'Bob'])

# 3. Join using index
joined_df = left.join(right)
print("\nJoined DataFrame:\n", joined_df)
```
