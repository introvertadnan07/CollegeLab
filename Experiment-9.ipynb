```python
# Calculator class
class Calculator:
    def add(self, a=0, b=0, c=0):
        return a + b + c


calc = Calculator()

print("Add two numbers:", calc.add(10, 20))
print("Add three numbers:", calc.add(10, 20, 14))
print("Add one number:", calc.add(20))


# Operator Overloading (+ operator)
class Student:
    def __init__(self, marks):
        self.marks = marks

    def __add__(self, other):
        return Student(self.marks + other.marks)

    def display(self):
        print("Total marks:", self.marks)


# Create objects
s1 = Student(77)
s2 = Student(87)

# Add objects using overloaded +
s3 = s1 + s2

# Display result
s3.display()
```
