```python
import re

# Sample string
text = "My email is adana.doe@example.com and my backup is mqa@sample.org"

# 1. Find all email addresses
emails = re.findall(r"\b[\w.-]+@[\w.-]+\.\w+\b", text)
print("Emails found:", emails)

# 2. Search for the word "backup"
match = re.search(r'backup', text)
if match:
    print("Found 'backup' at position:", match.start())
else:
    print("'backup' not found")

# 3. Match a pattern at the beginning
pattern = re.match(r"My email", text)
print("Match at beginning:", bool(pattern))

# 4. Replace email addresses with [EMAIL]
masked_text = re.sub(r"\b[\w.-]+@[\w.-]+\.\w+\b", "[EMAIL]", text)
print("Masked text:", masked_text)

# 5. Split text by spaces or punctuation
split_words = re.split(r"[\s,.]+", text)
print("Split words:", split_words)
```
